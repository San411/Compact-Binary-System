```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
path = '/content/drive/MyDrive/Internship/SSERD/Progress'
df = pd.read_csv(os.path.join(path, 'data.csv'))
```

```
df.head(25)
```

```
df["q"] = df["Secondary Mass"]/df["Primary Mass"]
```

```
df
```

```
df["eta"] = df["q"]/pow((1+ df["q"]),2)
```

```
df
```

```
global Am, Bm, Cm, chi
Am = 39.4112
Bm = 36.464
Cm = -115.811
# chi = -0.04

df["mf"] = Am + (Bm * (df["eta"] ** 2)) + (Cm * df['chi'])
df
```

```
plt.title('Mass vs Effective Spin')
plt.rc_context({'axes.edgecolor':'#581845', 'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'#F1D8DD'})
plt.xlabel('Mass')
plt.ylabel('Chi')
plt.scatter(df['mf(detected)'], df['chi'], c = '#F2BFDF')
plt.plot(df['mf'],df['chi'] )
plt.grid(True)
plt.show()
```

```
from sklearn.metrics import r2_score
r2_score(df['mf(detected)'],df['mf'])
```

```
plt.title("Detected Mass vs Predicted Mass",fontweight = 'bold')
plt.xlabel("",fontweight = 'bold')
plt.ylabel("mf(predicted)",fontweight = 'bold')
sns.scatterplot(x = 'mf(detected)',y = 'mf', data = df)
plt.rc_context({'axes.edgecolor':'#581845', 'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'#F1D8DD'})
plt.grid(True)
plt.show()
```

```
global Aa, Ba, Ca
Aa = 2.625528
Ba = 0.67426
Ca = 0.00071022
df["af"] = (Aa * df["eta"]) + (Ba * df["chi"]) - (Ca * (1 - df['mf']))
df
```

```
plt.title("Detected Spin vs Experimental Spin",fontweight = 'bold')
plt.xlabel("",fontweight = 'bold')
plt.ylabel("af(predicted)",fontweight = 'bold')
sns.scatterplot(x = 'af(detected)',y = 'af',data = df)
plt.rc_context({'axes.edgecolor':'#581845', 'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'#F1D8DD'})
plt.grid(True)
plt.show()
```

```
plt.title('Spin vs Effective Spin')
plt.rc_context({'axes.edgecolor':'#581845', 'xtick.color':'black', 'ytick.color':'black', 'figure.facecolor':'#F1D8DD'})
plt.xlabel('Spin')
plt.ylabel('Chi')
plt.scatter(df['af(detected)'], df['chi'], c = '#A96BA8')
plt.plot(df['af'],df['chi'] )
plt.grid(True)
plt.show()
```

```
from sklearn.metrics import r2_score
r2_score(df['af(detected)'],df['af'])
```
